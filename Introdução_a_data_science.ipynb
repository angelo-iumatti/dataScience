```python
import pandas as pd
filmes = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/ml-latest-small/movies.csv")
filmes.columns = ["filmeId", "titulo", "generos"] #comando para renomear as colunas
filmes.head() #aqui é o comando para mostrar o top 5
```

```python
notas.query("filmeId==1") #quero saber apenas as notas do filmeId 01 utiliza-se a palavra-chave QUERY
```

```python
#agora quero saber a média das notas do filme 01, para isso uso o comandon MEAN, porém se colocar apenas o MEAN vai aparecer todas as médias de todas as colunas
#por isso precio estabelecer qual coluna eu quero a média
notas.query("filmeId==1").mean()#aqui vai aperecer a média de todas as colunas
notas.query("filmeId==1")["nota"].mean()#aqui vai aparecer apenas a média da coluna nota
```

```python
notas.query("filmeId==2").nota.mean()#outro exemplo para saber a média das notas do filme 2
#o query é uma ferramenta muito poderosa para fazer pesquisas, buscar dados.
```

```python
#Agora vamos fazer a média das notas de todos os filmes por filme
#para isso utiliza-se o GROUP BY
notas.groupby("filmeId").mean()#aqui vai ser feito a média das notas de todos os filmes, com o uso do GROUP BY, porém vai ser feito tb das outras colunas
#para resolver isso utiliza-se a definição de qual coluna
notas.groupby("filmeId")["nota"].mean()
```

```python
#atendendo as boas praticas, atribuimos a uma variavel.
medias_por_filme = notas.groupby("filmeId")["nota"].mean()
#agora podemos pedir para visualizar
medias_por_filme
#ou então apenas as 5 primeiras utilizando o HEAD
medias_por_filme.head()
```

```python
#Agora podemos utilizar um gráfico de distribuição o PLOT e o tipo HIST
medias_por_filme.plot(kind='hist')
```

O método query() do Pandas permite filtrar dados de um DataFrame utilizando uma expressão de consulta semelhante à linguagem SQL
